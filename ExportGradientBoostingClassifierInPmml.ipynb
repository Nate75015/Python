```python
# IMPORT Modules
import pandas as pd 
import numpy as np 
import sklearn2pmml 
import sklearn_pandas 
import csv


from sklearn2pmml import sklearn2pmml
from sklearn2pmml import PMMLPipeline
from sklearn.ensemble import GradientBoostingClassifier 
```

```python
# load and prepare data
Mydata = pd.read_csv ("winequality-red.csv", sep = ";" ) 
Mydata.head() 
Mydata.shape 
Mydata.columns = ['FIXED_ACIDITY','VOLATILE_ACIDITY','CITRIC_ACID','RESIDUAL_SUGAR','CHLORIDES','FREE_SULFUR_DIOXIDE','TOTAL_SULFUR_DIOXIDE','DENSITY','PH','SULPHATES','ALCOHOL','QUALITY']
```

```python
Mydata1 = Mydata[:700]
Mydata2 = Mydata[700:1000]
```

```python
# Fit the model 
model = GradientBoostingClassifier() 


Target = Mydata1['QUALITY']
del Mydata1['QUALITY']
GradientBoostingClassifierPython = model.fit(Mydata1, Target)
GradientBoostingClassifierPython
```

```python
sklearn2pmml(PMMLPipeline([("classifier", GradientBoostingClassifierPython)]), "GradientBoostingClassifier.pmml", with_repr = True)
```

```python
import os, sys

pmml = open('GradientBoostingClassifier.pmml', 'r').read()
variables = list(Mydata1)
cnt = 1
for variable in variables:
    pmml = pmml.replace('"x' + str(cnt) + '"', '"' + variable + '"')
    cnt += 1
pmml = pmml.replace('"x' + str(cnt) + '"', '"' + variable + '"')
with open('GradientBoostingClassifier.pmml', 'w') as f:
    f.write(pmml)
```
